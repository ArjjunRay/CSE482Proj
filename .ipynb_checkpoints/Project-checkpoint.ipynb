```python
def get_json(url):
    r=requests.get(url, headers=HEADERS, timeout=60)
    if r.status_code!=200:
        raise RuntimeError(f"Request failed ({r.status_code}).\nURL: {url}\n{r.text[:300]}")
    return r.json()

def save_raw(obj, filename):
    path=os.path.join("data/raw", filename)
    with open(path,"w",encoding="utf-8") as f:
        json.dump(obj,f,indent=2)
    return path
```

```python
all_races = []
for season in SEASONS:
    url=f"{BASE}/races/{season}"
    print("Fetching races for", season)
    races=get_json(url)
    save_raw(races, f"races_{season}.json")
    all_races+=races
    time.sleep(SLEEP_SEC)

races_df=pd.DataFrame(all_races)
print("Raw races rows:", len(races_df), "cols:", len(races_df.columns))

wc=["RaceID", "SeriesID", "SeriesName", "Season", "Name", "Day", "DateTime", "Track", "IsInProgress", "IsOver"]
keep_cols=[c for c in wc if c in races_df.columns]
races_df=races_df[keep_cols].copy()
if "DateTime" in races_df.columns:
    races_df["DateTime"] = pd.to_datetime(races_df["DateTime"], errors="coerce")

if "IsOver" in races_df.columns:
    races_df=races_df[races_df["IsOver"]==True].copy()
races_df=races_df.dropna(subset=["RaceID"]).drop_duplicates(subset=["RaceID"])
out_path = "data/processed/races.csv"
races_df.to_csv(out_path, index=False)
print("Saved:",out_path,"rows=",len(races_df))
```

```python
drivers_url = f"{BASE}/Drivers"
print("Fetching drivers...")
drivers = get_json(drivers_url)
save_raw(drivers, "drivers.json")
drivers_df = pd.DataFrame(drivers)
wc=["DriverID", "FirstName", "LastName", "Number", "Team", "Manufacturer"]
keep=[c for c in wc if c in drivers_df.columns]
drivers_df = drivers_df[keep].drop_duplicates(subset=["DriverID"])
out_path = "data/processed/drivers.csv"
drivers_df.to_csv(out_path, index=False)
print("Saved:", out_path, "rows=", len(drivers_df))
```

```python
races_final = pd.read_csv("data/processed/races.csv")
race_ids = races_final["RaceID"].dropna().astype(int).tolist()
all_rows = []
failed = []
for rid in race_ids:
    url=f"{BASE}/raceresultfinal/{rid}"
    try:
        rr = get_json(url)
        save_raw(rr, f"raceresultfinal_{rid}.json")
        race_obj = rr.get("Race", {}) if isinstance(rr, dict) else {}
        driver_races = rr.get("DriverRaces", []) if isinstance(rr, dict) else []
        for dr in driver_races:
            all_rows.append({
                "RaceID": race_obj.get("RaceID", rid),
                "Season": race_obj.get("Season"),
                "SeriesName": race_obj.get("SeriesName"),
                "DateTime": race_obj.get("DateTime"),
                "Track": race_obj.get("Track"),
                "DriverID": dr.get("DriverID"),
                "StartPosition": dr.get("StartPosition") or dr.get("StartingPosition"),
                "FinishPosition": dr.get("FinishPosition") or dr.get("FinalPosition"),
                "Laps": dr.get("Laps") or dr.get("LapsCompleted"),
                "Points": dr.get("Points"),
            })

        time.sleep(SLEEP_SEC)

    except Exception as e:
        failed.append((rid, str(e)))

results_df = pd.DataFrame(all_rows)
print("Raw results rows:", len(results_df))
results_df["DateTime"] = pd.to_datetime(results_df["DateTime"], errors="coerce")
num_cols = ["RaceID", "DriverID", "StartPosition", "FinishPosition", "Laps", "Points"]
for col in num_cols:
    if col in results_df.columns:
        results_df[col] = pd.to_numeric(results_df[col], errors="coerce")
results_df = results_df.dropna(subset=["RaceID", "DriverID", "FinishPosition"])
results_df = results_df.drop_duplicates(subset=["RaceID", "DriverID"])
results_df["Top10"] = (results_df["FinishPosition"] <= 10).astype(int)

out_path = "data/processed/race_results.csv"
results_df.to_csv(out_path, index=False)

print("Saved:", out_path, "rows=", len(results_df))
print("Failed races:", len(failed))
if failed:
    print("First failure:", failed[0])
```

We retrieved the data using the API and stored it in pandas DataFrames. From the retrieved data, a clean and structured data set was formed by combining the necessary data. This was done by matching the race columns and ensuring the data sets were compatible.

The original data set had missing information and a few columns that were not necessary for the model’s predictions. The unwanted attributes were removed from the data set. The columns were also made uniform in name and data type for a smooth transition in the next phase.

New columns were added to the data set for easier analysis. For instance, new columns were added for easier classification. The data was also reorganized so that each row corresponds to a driver and a race.

This resulted in a clean and tabular data set ready for machine learning. The clean data set was ready for the next phase in the process.

```python
results = pd.read_csv("data/processed/race_results.csv")
drivers = pd.read_csv("data/processed/drivers.csv")
races = pd.read_csv("data/processed/races.csv")

results["DateTime"] = pd.to_datetime(results["DateTime"], errors="coerce")
races["DateTime"] = pd.to_datetime(races["DateTime"], errors="coerce")

for c in ["RaceID", "DriverID", "StartPosition", "FinishPosition", "Laps", "Points"]:
    if c in results.columns:
        results[c] = pd.to_numeric(results[c], errors="coerce")

drivers["DriverName"] = (drivers["FirstName"].fillna("") + " " + drivers["LastName"].fillna("")).str.strip()
race_info = races[["RaceID", "Name", "Track", "DateTime", "Season"]].rename(columns={"Name": "RaceName"})
driver_info = drivers[["DriverID", "DriverName", "Team", "Manufacturer"]]

df = results.merge(race_info, on="RaceID", how="left", suffixes=("", "_race"))
df = df.merge(driver_info, on="DriverID", how="left")
df["Track"] = df["Track_race"].combine_first(df["Track"])
df["DateTime"] = df["DateTime_race"].combine_first(df["DateTime"])
df["Season"] = df["Season_race"].combine_first(df["Season"])
df = df.drop(columns=[c for c in ["Track_race", "DateTime_race", "Season_race"] if c in df.columns])
df = df.dropna(subset=["RaceID", "DriverID", "StartPosition", "FinishPosition"])
df = df.drop_duplicates(["RaceID", "DriverID"]).sort_values(["DateTime", "RaceID", "DriverID"]).reset_index(drop=True)

df.head()
```

### Feature Selection

```python
def make_features(df):
    hist = defaultdict(list)
    track_hist = defaultdict(list)
    recent = defaultdict(lambda: deque(maxlen=5))
    rows = []

    for _, r in df.sort_values(["DateTime", "RaceID", "DriverID"]).iterrows():
        d = int(r["DriverID"])
        t = r["Track"] if pd.notna(r["Track"]) else "Unknown"
        prev = hist[d]
        prev_track = track_hist[(d, t)]

        rows.append({
            **r.to_dict(),
            "prior_races": len(prev),
            "prior_avg_finish": np.mean([x["finish"] for x in prev]) if prev else np.nan,
            "prior_avg_start": np.mean([x["start"] for x in prev]) if prev else np.nan,
            "prior_top10_rate": np.mean([x["finish"] <= 10 for x in prev]) if prev else np.nan,
            "prior_avg_points": np.mean([x["points"] for x in prev if pd.notna(x["points"])]) if prev else np.nan,
            "recent5_avg_finish": np.mean(recent[d]) if len(recent[d]) else np.nan,
            "track_prior_races": len(prev_track),
            "track_avg_finish": np.mean(prev_track) if prev_track else np.nan,
        })
        prev.append({
            "finish": float(r["FinishPosition"]),
            "start": float(r["StartPosition"]),
            "points": float(r["Points"]) if pd.notna(r["Points"]) else np.nan,
        })
        track_hist[(d, t)].append(float(r["FinishPosition"]))
        recent[d].append(float(r["FinishPosition"]))

    return pd.DataFrame(rows)

feat = make_features(df)
feat = feat[feat["prior_races"] >= 5].copy()

race_order = feat[["RaceID", "DateTime"]].drop_duplicates().sort_values(["DateTime", "RaceID"]).reset_index(drop=True)
split = int(len(race_order) * 0.8)
train_ids = set(race_order.iloc[:split]["RaceID"])
test_ids = set(race_order.iloc[split:]["RaceID"])

train = feat[feat["RaceID"].isin(train_ids)].copy()
test = feat[feat["RaceID"].isin(test_ids)].copy()
y_train = train["FinishPosition"]
y_test = test["FinishPosition"]
cols = ["StartPosition", "prior_races", "prior_avg_finish", "prior_avg_start","prior_top10_rate", "prior_avg_points","recent5_avg_finish", "track_prior_races", "track_avg_finish"]
lin_full = Pipeline([("imp", SimpleImputer(strategy="median")),("model", LinearRegression())])
tree_full = Pipeline([("imp", SimpleImputer(strategy="median")),("model", DecisionTreeRegressor(max_depth=8, min_samples_leaf=5, random_state=42))])
lin_full.fit(train[cols], y_train)
tree_full.fit(train[cols], y_train)

pred_lin = lin_full.predict(test[cols])
pred_tree = tree_full.predict(test[cols])
print("train rows:", len(train), "test rows:", len(test))
feat[cols + ["FinishPosition"]].head()
```

### Given a set of drivers participating in the same race, predict which driver will finish ahead

```python
def make_pairs(x):
    rows = []
    for race_id, g in x.groupby("RaceID"):
        recs = g[["RaceID", "FinishPosition"] + cols].to_dict("records")
        for a, b in combinations(recs, 2):
            row = {"RaceID": race_id, "label": int(a["FinishPosition"] < b["FinishPosition"])}
            for c in cols:
                row[c + "_diff"] = a[c] - b[c] if pd.notna(a[c]) and pd.notna(b[c]) else np.nan
            rows.append(row)
    return pd.DataFrame(rows)

pairs = make_pairs(feat)
train_pairs = pairs[pairs["RaceID"].isin(train_ids)].copy()
test_pairs = pairs[pairs["RaceID"].isin(test_ids)].copy()
pair_cols = [c + "_diff" for c in cols]

logit = Pipeline([("imp", SimpleImputer(strategy="median")),("scale", StandardScaler()),("model", LogisticRegression(max_iter=1000))])
ctree = Pipeline([("imp", SimpleImputer(strategy="median")),("model", DecisionTreeClassifier(max_depth=8, min_samples_leaf=5, random_state=42))])
logit.fit(train_pairs[pair_cols], train_pairs["label"])
ctree.fit(train_pairs[pair_cols], train_pairs["label"])

pred_logit = logit.predict(test_pairs[pair_cols])
prob_logit = logit.predict_proba(test_pairs[pair_cols])[:, 1]
pred_ctree = ctree.predict(test_pairs[pair_cols])
prob_ctree = ctree.predict_proba(test_pairs[pair_cols])[:, 1]

pair_results = pd.DataFrame([["LogisticRegression", accuracy_score(test_pairs["label"], pred_logit), roc_auc_score(test_pairs["label"], prob_logit)],["DecisionTree", accuracy_score(test_pairs["label"], pred_ctree), roc_auc_score(test_pairs["label"], prob_ctree)],], columns=["model", "accuracy", "roc_auc"]).sort_values("accuracy", ascending=False)

print("Results for finishing ahead")
display(pair_results.round(4))
```

### Given a set of drivers starting from different starting positions, will the starting affect race-end psoitions.

```python
spearman_all = feat["StartPosition"].corr(feat["FinishPosition"], method="spearman")
spearman_test = test["StartPosition"].corr(test["FinishPosition"], method="spearman")
start_train = train[["StartPosition"]]
start_test = test[["StartPosition"]]
lin_start = Pipeline([("imp", SimpleImputer(strategy="median")),("model", LinearRegression())])

tree_start = Pipeline([("imp", SimpleImputer(strategy="median")),("model", DecisionTreeRegressor(max_depth=5, min_samples_leaf=5, random_state=42))])

lin_start.fit(start_train, y_train)
tree_start.fit(start_train, y_train)

pred_lin_start = lin_start.predict(start_test)
pred_tree_start = tree_start.predict(start_test)

no_start_cols = [c for c in cols if c != "StartPosition"]

lin_no_start = Pipeline([("imp", SimpleImputer(strategy="median")),("model", LinearRegression())])

tree_no_start = Pipeline([("imp", SimpleImputer(strategy="median")),("model", DecisionTreeRegressor(max_depth=8, min_samples_leaf=5, random_state=42))])

lin_no_start.fit(train[no_start_cols], y_train)
tree_no_start.fit(train[no_start_cols], y_train)

pred_no_start_lin = lin_no_start.predict(test[no_start_cols])
pred_no_start_tree = tree_no_start.predict(test[no_start_cols])

start_results = pd.DataFrame([
    ["Spearman corr (all)", spearman_all],
    ["Spearman corr (test)", spearman_test],
    ["Linear start-only MAE", mean_absolute_error(y_test, pred_lin_start)],
    ["Tree start-only MAE", mean_absolute_error(y_test, pred_tree_start)],
    ["Linear full model MAE", mean_absolute_error(y_test, pred_lin)],
    ["Linear without StartPosition MAE", mean_absolute_error(y_test, pred_no_start_lin)],
    ["Tree full model MAE", mean_absolute_error(y_test, pred_tree)],
    ["Tree without StartPosition MAE", mean_absolute_error(y_test, pred_no_start_tree)],
], columns=["metric", "value"])

display(start_results.round(4))
```

```python
sample = test.sample(min(len(test), 300), random_state=42)

plt.figure(figsize=(7, 4))
plt.scatter(sample["StartPosition"], sample["FinishPosition"], alpha=0.35)

m, b = np.polyfit(sample["StartPosition"], sample["FinishPosition"], 1)
x = np.linspace(sample["StartPosition"].min(), sample["StartPosition"].max(), 100)
plt.plot(x, m * x + b)

plt.xlabel("Start Position")
plt.ylabel("Finish Position")
plt.title("Start Position vs Finish Position")
plt.show()
```

```python
start_bins = test.copy()
start_bins["start_bin"] = pd.cut(start_bins["StartPosition"], bins=8)

bin_means = start_bins.groupby("start_bin", observed=False)["FinishPosition"].mean()

plt.figure(figsize=(8, 4))
plt.plot(range(len(bin_means)), bin_means.values, marker="o")
plt.xticks(range(len(bin_means)), [str(x) for x in bin_means.index], rotation=30)
plt.xlabel("Start Position Bin")
plt.ylabel("Average Finish Position")
plt.title("Average Finish Position by Starting Bin")
plt.show()
```

We compared several models to predict race outcomes. For finishing positions, Linear Regression performed best, giving the lowest error on test races. For predicting which driver finishes ahead, Logistic Regression worked best, with about 68% accuracy and a ROC-AUC of around 0.75. Overall, the models were able to pick up useful patterns from past race data and make reasonably accurate predictions.

### Does recent form matter more than long term average?

```python
recent_cols = ["StartPosition", "prior_races", "recent5_avg_finish"]
career_cols = ["StartPosition", "prior_races", "prior_avg_finish"]
both_cols = ["StartPosition", "prior_races", "recent5_avg_finish", "prior_avg_finish"]
feature_sets = {
    "recent_only": recent_cols,
    "career_only": career_cols,
    "recent_plus_career": both_cols
}
recent_results = []
for name, use_cols in feature_sets.items():
    X_train = train[use_cols]
    X_test = test[use_cols]
    y_train = train["FinishPosition"]
    y_test = test["FinishPosition"]
    lin_model = Pipeline([("imp", SimpleImputer(strategy="median")),("model", LinearRegression())])

    tree_model = Pipeline([("imp", SimpleImputer(strategy="median")),("model", DecisionTreeRegressor(max_depth=8, min_samples_leaf=5, random_state=42))])
    lin_model.fit(X_train, y_train)
    tree_model.fit(X_train, y_train)
    pred_lin = lin_model.predict(X_test)
    pred_tree = tree_model.predict(X_test)
    recent_results.append([name,"LinearRegression",mean_absolute_error(y_test, pred_lin),math.sqrt(mean_squared_error(y_test, pred_lin))])

    recent_results.append([name,"DecisionTree",mean_absolute_error(y_test, pred_tree),math.sqrt(mean_squared_error(y_test, pred_tree))])
recent_results = pd.DataFrame(recent_results,columns=["feature_set", "model", "mae", "rmse"]).sort_values("mae")
display(recent_results.round(4))
```

```python
sample_recent = test.dropna(subset=["recent5_avg_finish", "FinishPosition"]).sample(
    min(300, test.dropna(subset=["recent5_avg_finish", "FinishPosition"]).shape[0]),
    random_state=42)

plt.figure(figsize=(7, 4))
plt.scatter(sample_recent["recent5_avg_finish"], sample_recent["FinishPosition"], alpha=0.35)

m, b = np.polyfit(sample_recent["recent5_avg_finish"], sample_recent["FinishPosition"], 1)
x = np.linspace(sample_recent["recent5_avg_finish"].min(), sample_recent["recent5_avg_finish"].max(), 100)
plt.plot(x, m * x + b)

plt.xlabel("Recent Average Finish (Last 5 Races)")
plt.ylabel("Finish Position")
plt.title("Recent Form vs Finish Position")
plt.show()
```

```python
sample_career = test.dropna(subset=["prior_avg_finish", "FinishPosition"]).sample(
    min(300, test.dropna(subset=["prior_avg_finish", "FinishPosition"]).shape[0]),
    random_state=42)

plt.figure(figsize=(7, 4))
plt.scatter(sample_career["prior_avg_finish"], sample_career["FinishPosition"], alpha=0.35)

m, b = np.polyfit(sample_career["prior_avg_finish"], sample_career["FinishPosition"], 1)
x = np.linspace(sample_career["prior_avg_finish"].min(), sample_career["prior_avg_finish"].max(), 100)
plt.plot(x, m * x + b)

plt.xlabel("Career Average Finish Prior to Race")
plt.ylabel("Finish Position")
plt.title("Long-Term Average vs Finish Position")
plt.show()
```

### Does car driven affect driver position?

```python
base_cols = ["StartPosition", "prior_races", "prior_avg_finish", "recent5_avg_finish"]
manu_cols = base_cols + ["Manufacturer"]
car_results = []
for name, use_cols in {"without_manufacturer": base_cols, "with_manufacturer": manu_cols}.items():
    numeric_cols = [c for c in use_cols if c != "Manufacturer"]
    categorical_cols = [c for c in use_cols if c == "Manufacturer"]

    preprocessor = ColumnTransformer([("num", Pipeline([("imp", SimpleImputer(strategy="median"))]), numeric_cols),("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),("onehot", OneHotEncoder(handle_unknown="ignore"))]), categorical_cols)])
    lin_model = Pipeline([("prep", preprocessor),("model", LinearRegression())])

    tree_model = Pipeline([("prep", preprocessor),("model", DecisionTreeRegressor(max_depth=8, min_samples_leaf=5, random_state=42))])
    X_train = train[use_cols]
    X_test = test[use_cols]
    y_train = train["FinishPosition"]
    y_test = test["FinishPosition"]

    lin_model.fit(X_train, y_train)
    tree_model.fit(X_train, y_train)

    pred_lin = lin_model.predict(X_test)
    pred_tree = tree_model.predict(X_test)

    car_results.append([name,"LinearRegression",mean_absolute_error(y_test, pred_lin),math.sqrt(mean_squared_error(y_test, pred_lin))])
    car_results.append([name,"DecisionTree",mean_absolute_error(y_test, pred_tree),math.sqrt(mean_squared_error(y_test, pred_tree))])
car_results = pd.DataFrame(car_results,columns=["feature_set", "model", "mae", "rmse"]).sort_values("mae")
display(car_results.round(4))
```

```python
manu_plot = feat.dropna(subset=["Manufacturer", "FinishPosition"]).copy()

top_manus = manu_plot["Manufacturer"].value_counts().head(8).index
manu_plot = manu_plot[manu_plot["Manufacturer"].isin(top_manus)]

manu_means = manu_plot.groupby("Manufacturer")["FinishPosition"].mean().sort_values()

plt.figure(figsize=(8, 4))
plt.bar(manu_means.index, manu_means.values)
plt.xlabel("Manufacturer")
plt.ylabel("Average Finish Position")
plt.title("Average Finish Position by Manufacturer")
plt.xticks(rotation=30)
plt.show()
```

```python
manu_top10 = manu_plot.groupby("Manufacturer")["FinishPosition"].apply(lambda x: (x <= 10).mean()).sort_values(ascending=False)
plt.figure(figsize=(8, 4))
plt.bar(manu_top10.index, manu_top10.values)
plt.xlabel("Manufacturer")
plt.ylabel("Top-10 Rate")
plt.title("Top-10 Rate by Manufacturer")
plt.xticks(rotation=30)
plt.show()
```

We also looked at whether recent form and car type affect performance. For recent vs long-term form, Linear Regression showed that recent performance slightly outperformed long-term averages, but the difference was small, and combining both did not improve results much. This suggests recent form matters a bit more, but both contain similar information. For car manufacturer, adding it to the model did not improve performance, as errors stayed almost the same with and without it. Overall, recent form has a small impact, while car type does not seem to significantly affect finishing position.
